```python
from influxdb_interface import Influx
import influxdb
import utility as util
from datetime import datetime
import pandas as pd
```

```python
df = pd.read_csv('book_with_grids_combined.csv')

# get info from Influx
db = Influx()
client = db.get_client()
```

```python
# get measurements
measurements = client.get_list_measurements()
measurements
```

```python
#  create start_date and end_date: April 25, 2022 to July 20, 2022
start_date = datetime(2022, 4, 25, 0, 0, 0)
end_date = datetime(2022, 7, 20, 0, 0, 0)
```

```python
# get csv of participant awair sensors
pids = pd.read_csv('wb_awairs.csv')
```

```python
# create type column with value 'awair_omni'
pids['type'] = 'awair_omni'
pids # awairs installed off grounds are not in the book_with_grids_combined.csv
```

```python
score = util.get_lfdf("awair_score", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
temp = util.get_lfdf("Temperature_°C", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
humid = util.get_lfdf("Humidity_%", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
co2 = util.get_lfdf("co2_ppm", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
voc = util.get_lfdf("voc_ppb", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
lum = util.get_lfdf("Illumination_lx", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
noise = util.get_lfdf("spl_a", start_date, end_date, list(pids[pids['type']=='awair_omni']['device_id']))
```

```python
# remove columns except for time, device_id, and value for score, co2, voc, lum
all_score = score[['time', 'device_id', 'value']]
# rename value to score
all_score = all_score.rename(columns={'value': 'score'})

all_temp = temp[['time', 'device_id', 'value']]
# rename value
all_temp = all_temp.rename(columns={'value': 'temp'})

all_humid = humid[['time', 'device_id', 'value']]
# rename value
all_humid = all_humid.rename(columns={'value': 'humid'})

all_co2 = co2[['time', 'device_id', 'value']]
# rename value
all_co2 = all_co2.rename(columns={'value': 'co2'})

all_voc = voc[['time', 'device_id', 'value']]
# rename value
all_voc = all_voc.rename(columns={'value': 'voc'})

all_lum = lum[['time', 'device_id', 'value']]
# rename value
all_lum = all_lum.rename(columns={'value': 'lum'})

all_noise = noise[['time', 'device_id', 'value']]
# rename value
all_noise = all_noise.rename(columns={'value': 'noise'})
```

```python
# merge all dataframes on device_id and time
all_data = pd.merge(all_score, all_temp, on=['device_id', 'time'])
all_data = pd.merge(all_data, all_humid, on=['device_id', 'time'])
all_data = pd.merge(all_data, all_co2, on=['device_id', 'time'])
all_data = pd.merge(all_data, all_voc, on=['device_id', 'time'])
all_data = pd.merge(all_data, all_lum, on=['device_id', 'time'])
all_data = pd.merge(all_data, all_noise, on=['device_id', 'time'])
```

```python
# add participant_id column to all_data by merging on device_id
all_data = pd.merge(all_data, pids, on='device_id')
# drop columns Awair, Work, type
all_data = all_data.drop(columns=['Awair', 'Work', 'type'])
all_data.head()
```

```python
# group data by Participant ID and sort by time
all_data = all_data.groupby(['ParticipantID']).apply(lambda x: x.sort_values(['time']))
```

```python
all_data.head()
```

```python
# create csv for each unique ParticipantID
for pid in all_data['ParticipantID'].unique():
    # get data for pid
    pid_data = all_data[all_data['ParticipantID']==pid]
    # write to csv
    pid_data.to_csv('lll_awair/{}_awair.csv'.format(pid), index=False)

```
